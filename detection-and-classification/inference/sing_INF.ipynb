```python
from ultralytics import YOLO
import cv2

# Charger le modèle entraîné (ADAPTE ce chemin)
model = YOLO("/home/imadb/runs/detect/RoadWorks2/weights/best.pt")

# Image à tester (ADAPTE ce chemin)
img_path = "/home/imadb/Ghiles/road work sing 1/road works sign/test/images/-2-_jpeg.rf.2e2dfce4ed330ba62b99f2272a962384.jpg"

# Inference
results = model(img_path, verbose=False)[0]

print("=== Détections RoadWorks ===")
for box in results.boxes:
    cls = int(box.cls)
    name = results.names[cls]
    conf = float(box.conf)
    print(f"- {name} ({conf:.2f})")

# Dessin
img = cv2.imread(img_path)

for box in results.boxes:
    x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
    cls = int(box.cls)
    name = results.names[cls]
    conf = float(box.conf)

    label = f"{name} {conf:.2f}"
    color = (0, 255, 0)

    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img, label, (x1, max(15, y1 - 5)),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2, cv2.LINE_AA)

out_path = "result_roadworks.jpg"
cv2.imwrite(out_path, img)
print("Image sauvegardée :", out_path)
```
