```python
from ultralytics import YOLO
import cv2
import os

# 1) On charge le modèle entraîné Road Debris 1
model = YOLO("/home/imadb/runs/detect/RoadDebris1/weights/best.pt")

# 2) Image à tester (une image du dossier test)
img_path = (
    "/home/imadb/Ghiles/debris1/Road debrit 1/test/images/182_jpeg.rf.61288886b17d065029248ece304aca71.jpg"
)

# Vérification
assert os.path.exists(img_path), f"Image non trouvée : {img_path}"

# 3) Inference 
results = model(img_path, conf=0.25, device=0, verbose=False)[0]

print("=== Détections ===")
if len(results.boxes) == 0:
    print("Aucun debris détecté")

for box in results.boxes:
    cls = int(box.cls)
    name = results.names[cls]
    conf = float(box.conf)
    print(f"- {name} ({conf:.2f})")

img = cv2.imread(img_path)
# 6) Sauvegarder le résultat
out_path = "result_road_debris1.jpg"
cv2.imwrite(out_path, img)

print("Image sauvegardée :", out_path)
```
